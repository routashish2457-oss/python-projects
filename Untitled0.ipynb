```python
try:
    value = input("Enter a number or word: ")
    number = int(value)
    if number % 2 == 0:
        print(f"{number} is an even number.")
    else:
        print(f"{number} is an odd number.")
except ValueError:
    print(f"'{value}' is not a valid number. Please enter a number.")
```
